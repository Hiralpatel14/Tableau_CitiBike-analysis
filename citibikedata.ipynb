```python
# Add dependencies
import glob
import pandas as pd
```

```python
# glob.glob('data*.csv') - returns List[str]
# pd.read_csv(f) - returns pd.DataFrame()
# for f in glob.glob() - returns a List[DataFrames]
# pd.concat() - returns one pd.DataFrame()
clean_ALLdata_trip = pd.concat([pd.read_csv(f) for f in glob.glob('2017allmonthsdata/*.csv')], ignore_index = True, sort=False)
clean_QTRdata_rider = pd.concat([pd.read_csv(f) for f in glob.glob('Data2017-QTR/*.csv')], ignore_index = True, sort=True)
```
